```python
from vggTrimmedModel import TrimmedModel
from CIFAR_DataLoader import CifarDataManager
import numpy as np
```

```python
data_loader = CifarDataManager()
```

## 统计multi-class的accuracy平均数, num of classes:25

```python
average_random = 0
for _ in range(50):
    # average = 0
    target_class_id = np.random.randint(100,size=25)
    print("----------------------------------------------------------------------------------------------")
    model = TrimmedModel(target_class_id=target_class_id, multiPruning=True)
    for _ in range(100):
        test_images, test_labels = data_loader.test.next_batch(1000)
        model.assign_weight()
        accuracy = model.test_accuracy(test_images, test_labels)
        # average += accuracy
        break
    print(str(target_class_id))
    print("Test Accuracy: " + str(accuracy))
    average_random += accuracy
print("AVERAGE ACCURACY FOR" + str(50)+"-CLASS: ", average_random/50)
```
